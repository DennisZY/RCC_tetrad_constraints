```python
import pandas as pd
import seaborn as sns
```

```python
real_df = pd.read_csv("real_data/poldem_values.csv")
real_df = real_df.iloc[:,0:300]

syn_df = pd.read_csv("generated_data.nosync/k1_d1_p1-0.5_p2-0.5_v0.5_gen_values.csv", names=list(range(300)))
```

```python
def four_vars(df, row):
    vars = [df.iloc[row,0:75],df.iloc[row,75:150],df.iloc[row,150:225],df.iloc[row,225:300]]
    for i, var in enumerate(vars):
        sns.distplot(var, hist = False, kde = True,
                 kde_kws = {'linewidth': 3},
                 label = i)
    
```

```python
four_vars(syn_df, 120)
```
