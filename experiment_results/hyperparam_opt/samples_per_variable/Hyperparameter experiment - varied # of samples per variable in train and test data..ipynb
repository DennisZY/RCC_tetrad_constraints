# Hyperparameter experiment - varied # of samples per variable in train and test data.

In this experiment the amount of samples per test and train set were varied. The assumption was made that it is best to have the same amount of samples per variable in the train and test set for the KME, so they where always set to the same value.

For each experiment 10 different test sets were generated.

```python
import pandas as pd
import numpy as np
```

```python
KME_results = pd.read_csv('experiment_sample_size_results.csv')
wishart_results = pd.read_csv('wishart_experiment_samplesize_results.csv')
```

```python
for value in KME_results.n_samples.unique():
    print(value)
    print(KME_results[KME_results['n_samples'] == value]['mean_score'].mean())
    print(KME_results[KME_results['n_samples'] == value]['var_score'].mean())
```
