```python
import pandas as pd
import numpy as np
import tabulate
import matplotlib.pyplot as plt
import seaborn as sns
import math
import seaborn as sns
import itertools
```

```python
def sens_and_spec(df):
    sensitivity = df['truepos'] / (df['truepos'] + df['falseneg'])
    specificity = df['trueneg'] / (df['trueneg'] + df['falsepos'])

    sensitivity[np.isnan(sensitivity)] = 0
    specificity[np.isnan(specificity)] = 0
    return(sensitivity, specificity)

def KME_mean_pure(df):
    df_mean = df.groupby(['b','d','KME','E','K','n_samples','n_distributions']).mean()
    df_mean.insert(1, 'acc_std', df.groupby(['b','d','KME','E','K','n_samples','n_distributions'])[['score']].std())
    df_mean = df_mean.reset_index()
    df_mean['purity'] = True
    return(df_mean)

def KME_mean_impure(df):
    df_mean = df.groupby(['b','d','KME','E','K','n_samples','n_distributions']).mean()
    df_mean.insert(1, 'acc_std', df.groupby(['b','d','KME','E','K','n_samples','n_distributions'])[['score']].std())
    df_mean = df_mean.reset_index()
    df_mean['purity'] = False
    return(df_mean)

def wish_mean(df):
    df = df.drop(columns = ['model_count'])
    df = df.rename(columns={'alpha': 'p-value', 'linear': 'linearity'})
    df_mean = df.groupby(['linearity','p-value','b','d','n_samples']).mean()
    df_mean.insert(2, 'acc_std', df.groupby(['linearity','p-value','b','d','n_samples'])[['accuracy']].std())
    df_mean = df_mean.reset_index()
    return(df_mean)
    
```

```python
def plot_nsamples_sns(df):
    fig, axs = plt.subplots(2,2, sharex=True, sharey=True, figsize=(8,8))
    #fig.subplots_adjust(hspace = 1.2, wspace=1.2)
    
    palette = sns.color_palette("rocket", 3)

    axs = axs.ravel()

    for n, samples in enumerate(df['n_samples'].unique()):
        new_df = df[df['n_samples'] == samples]
        sens, spec = sens_and_spec(new_df)
        
        axs[n].plot([1, 0], [0, 1], linewidth=0.5)


        sns.scatterplot(x=spec, y=sens, ax=axs[n], palette = palette, s=80, hue=new_df['n_distributions'])
        axs[n].set_title('Sample size {}'.format(samples))
        axs[n].set(xlabel = 'Specificity', ylabel = 'Sensitivity')
        axs[n].set_xlim([-0.05,1.05])
        axs[n].set_ylim([-0.05,1.05])

        plt.xlabel('Specificity')
        plt.ylabel('Sensitivity')
        
def lin_vs_nonlin(lin_df, nonlin_df):
    fig, axs = plt.subplots(2,2, sharex=True, sharey=True, figsize=(10,10))
    #fig.subplots_adjust(hspace = 1.2, wspace=1.2)
    
    df = lin_df.append(nonlin_df)
    
    plt.style.use('fivethirtyeight')
    
    palette = sns.color_palette("rocket", 2)

    axs = axs.ravel()

    for n, samples in enumerate(df['n_samples'].unique()):
        new_df = df[df['n_samples'] == samples]
        sens, spec = sens_and_spec(new_df)
        
        axs[n].plot([1, 0], [0, 1], linewidth=0.5)


        sns.scatterplot(x=spec, y=sens, ax=axs[n], palette = palette, s=120, style = new_df['graph'],  hue=new_df['purity'])
        axs[n].set_title('Sample size {}'.format(samples))
        axs[n].set(xlabel = 'Specificity', ylabel = 'Sensitivity')
        axs[n].set_xlim([-0.05,1.05])
        axs[n].set_ylim([-0.05,1.05])

        plt.xlabel('Specificity')
        plt.ylabel('Sensitivity')
        
```

# KME VS WISHART

```python
results_1 = pd.read_csv('wishart/wishart_experiment_1_results.csv')
results_3 = pd.read_csv('wishart/wishart_experiment_3_results.csv')
results_cross = pd.read_csv('wishart/wishart_experiment_cross_construct_impure_results.csv')
results_intra = pd.read_csv('wishart/wishart_experiment_intra_construct_impure_results.csv')
results_latent = pd.read_csv('wishart/wishart_experiment_latent_measure_impure_results.csv')
```

```python
wish_1_mean = wish_mean(results_1)
wish_1_mean['graph'] = '1'

wish_3_mean = wish_mean(results_3)
wish_3_mean['graph'] = '3'

wish_cross_mean = wish_mean(results_cross)
wish_cross_mean['graph'] = 'cross'

wish_intra_mean = wish_mean(results_intra)
wish_intra_mean['graph'] = 'intra'

wish_latent_mean = wish_mean(results_latent)
wish_latent_mean['graph'] = 'latent'
```

```python
wish_impure_mean_all = wish_cross_mean.append([wish_intra_mean,wish_latent_mean])
wish_basic_mean_all = wish_1_mean.append([wish_3_mean])
```

```python
def wish_vs_kme(wish_df, kme_df , linearity):
    wish_df['test'] = 'Wishart'
    kme_df['test'] = 'KME'
    fig, axs = plt.subplots(2,2, sharex=True, sharey=True, figsize=(10,10))
    #fig.subplots_adjust(hspace = 1.2, wspace=1.2)
    
    wish_df = wish_df[wish_df['linearity'] == linearity]

    
    df = wish_df.append(kme_df)
        
    plt.style.use('fivethirtyeight')
    
    palette = sns.color_palette("rocket", 2)

    axs = axs.ravel()

    for n, samples in enumerate(df['n_samples'].unique()):
        new_df = df[df['n_samples'] == samples]
        sens, spec = sens_and_spec(new_df)
        
        axs[n].plot([1, 0], [0, 1], linewidth=0.5)


        sns.scatterplot(x=spec, y=sens, ax=axs[n], palette = palette, s=120, style = new_df['graph'],  hue=new_df['test'])
        axs[n].set_title('Sample size {}'.format(samples))
        axs[n].set(xlabel = 'Specificity', ylabel = 'Sensitivity')
        axs[n].set_xlim([-0.05,1.05])
        axs[n].set_ylim([-0.05,1.05])

        plt.xlabel('Specificity')
        plt.ylabel('Sensitivity')
```
