# Hyper parameter experiment - varied KME, E and K

In this experiment three variables were varied:
- KME
- E
- K

As the results demonstrate, the best scores where achieved with a KME that embeds all four variables at the same time (4) or one that embeds the two pairs that are being tested on t-separation. The KME that uses marginal distributions does not work.

Interesting to see is that the linearly generated data with 0 randomness is harder to classify than the data with 005 randomness. This corresponds to what Scholkopf says about the zero mean Gaussian case being what is most difficult to crack (I am however not sure how this exactly corresponds, but I presume it is related). 

```python
import pandas as pd
import numpy as np
import itertools
```

```python
results_0 = pd.read_csv('0_results.csv')
# Missing values here.
#results_001 = pd.read_csv('001_results.csv')
results_005 = pd.read_csv('005_results.csv')
results_ran_001 = pd.read_csv('001_ran_results.csv')
results_ran_005 = pd.read_csv('005_ran_results.csv')
```

```python
for i in [results_0, results_005, results_ran_001, results_ran_005]:
    df = i
    df['var_score'] = np.sqrt(df['var_score'])
    print(df[df['KME'] == '4'].describe()[['best_score','mean_score','var_score']])
```

```python
E_list = [500,1000,1500]
K_list = [200,400,800]
total = pd.concat([results_0, results_005, results_ran_001, results_ran_005])
total['var_score'] = np.sqrt(total['var_score'])
for prod in itertools.product(E_list,K_list):
    E, K = prod
    print(prod)
    print(total[(total['E'] == E) & (total['K'] == K)].describe()[['best_score','mean_score','var_score']])
```
